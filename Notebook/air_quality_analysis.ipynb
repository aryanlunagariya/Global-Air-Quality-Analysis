# DATA CLEANING & PREPROCESSING (Pandas + NumPy)

```python

import numpy as np
import pandas as pd

df = pd.read_csv("D:\Ds & Gen Ai\Global Air Quality Analysis\Data\globalAirQuality.csv")

# Handle missing values using appropriate techniques (mean/median/mode).
    # numerical columns → fill with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

    # categorical columns → fill with mode
cat_cols = df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Remove duplicate rows and justify the choice
df.duplicated().copy()

# Detect and handle outliers using IQR or Z-score method (NumPy based)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Create a new column categorizing air quality levels (Good, Moderate, Poor).
df['air_quality_level'] = df['aqi'].apply(
    lambda x: 'Good' if x <= 50 else ('Moderate' if x <= 100 else 'Poor'))

# Normalize selected numerical columns using Min-Max normalization.
cols = ['pm25', 'pm10', 'no2', 'so2', 'co', 'o3']
df[cols] = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
df[cols]

# Standardize selected columns using Z-score normalization.
cols = ['pm25', 'pm10', 'no2', 'so2', 'co', 'o3']
df[cols] = (df[cols] - df[cols].mean()) / df[cols].std()
df

# Filter data for a specific country or city using Pandas indexing.
df[df["country"]=='IN']
df[df['city'] == 'Delhi']

# Sort the dataset based on pollution level in descending order.
df_sorted = df.sort_values('aqi', ascending=False)
df_sorted
```

# VISUALIZATION QUESTIONS (Matplotlib + Seaborn)

```python
import matplotlib.pyplot as plt

 # Plot a line chart showing pollution trend over time.
df['timestamp'] = pd.to_datetime(df['timestamp'])

trend = (
    df.set_index('timestamp')['aqi']
      .resample('D')
      .mean()
)

plt.figure(figsize=(10, 4))
plt.plot(trend.index, trend.values, linewidth=1)
plt.title("pollution trend over time")
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Create a bar chart comparing average pollution levels across top 10 cities.
import matplotlib.pyplot as plt

avg_city = (
    df.groupby('city')['aqi']
      .mean()
      .sort_values(ascending=False)
      .head(10))

plt.figure(figsize=(10, 5))
plt.bar(avg_city.index, avg_city.values)
plt.title("Average polution")
plt.xlabel('City')
plt.ylabel('Average AQI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
